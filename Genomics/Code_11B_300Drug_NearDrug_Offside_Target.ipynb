```python
def mapping(df,dfkey,dfvalue):
    key=dfkey
    value=dfvalue
    dictionary={}

    for i in range(len(df)):
        dictionary[key[i]]=value[i]
        
    return dictionary
```

```python
s2=mapping(df,df['DRUG_NAME'].str.lower(),df['DRUG_ID'])
```

```python
df_Offside['DB_DRUG_ID'] = df_Offside['drugname'].str.lower().map(s2)
```

```python
df_Offside.isnull().sum()
```

```python
df_Offside.dropna(subset = ["DB_DRUG_ID"], inplace=True)
```
