```python
drugname=df["Drug_Name"]
target_name=df["Target_Name"]
```

```python
drugTarget={}
for i in range(len(df)):
    key=drugname[i].lower()
    value=target_name[i]
    drugTarget[key]=value
```

```python
df_s['Target_Name'] = df['Drug_Name'].str.lower().map(drugTarget)
df1= df_s[['DRUG_NAME','Target_Name']]
```
