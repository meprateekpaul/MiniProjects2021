```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
df = pd.read_csv('NEW_Table_3A_DrugScoreMatrix.csv')
length=len(df)
field_with_drug_name='drugname'
```

```python
df
```

```python
df.isnull().sum()
```

```python
dict={}
for i in tqdm(range(length)):
    
    drugnameC=df.columns[6+i]
    df_sort=df.sort_values(by=drugnameC, ascending=False)
    arrangedlist=df_sort[field_with_drug_name].to_numpy()
    dict[drugnameC]=arrangedlist[1]+";"+arrangedlist[2]+";"+arrangedlist[3]+";"+arrangedlist[4]+";"+arrangedlist[5]
```

```python
df_final=df_sort[['CID',field_with_drug_name,'no_of_side_effect','side_effect_name']]
```

```python
df_final['NearDrugs'] = df_final[field_with_drug_name].map(dict)
```
