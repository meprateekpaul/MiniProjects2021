```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
df = pd.read_csv('TTD_RAW_P1-07-Drug-TargetMapping.csv')
df=df.drop_duplicates().reset_index()
```

```python
def mapping(df,dfkey,dfvalue):
    key=dfkey
    value=dfvalue
    dictionary={}

    for i in range(len(df)):
        dictionary[key[i]]=value[i]
        
    return dictionary
```
