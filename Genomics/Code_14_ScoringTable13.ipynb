```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
def mapping(df,dfkey,dfvalue):
    key=dfkey
    value=dfvalue
    dictionary={}

    for i in range(len(df)):
        dictionary[key[i]]=value[i]
        
    return dictionary
```

```python
df = pd.read_csv('Table_13A_True_Negetive_Sider.csv')
length=len(df)
```

```python
df
```

```python
Drug1_Target_Name=df['Drug1_Target_Name']
Drug2_Target_Name=df['Drug2_Target_Name']

dct={}

for i in tqdm(range(len(df))):
    value = 0
    if Drug1_Target_Name[i]==Drug2_Target_Name[i]:
        value = 1
    else:
        value = 0
    
    dct[i]=value
```

```python
df['Matching_Score'] = df.index.to_series().map(dct)
```

```python
df
```

```python
df['Matching_Score'].value_counts()
```

```python
df = df.sort_values("Matching_Score", ascending = False)
```

```python
df.to_csv('Table_13A_True_Negetive_Sider_Scored.csv', index=False)
```
