```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
df = pd.read_csv('S2.csv')
length=len(df)
```

```python
df.drop_duplicates().reset_index()
df=df[['DRUG_NAME','DRUG_ID','SMILES']].drop_duplicates().reset_index()
```

```python
df=df[['DRUG_NAME','DRUG_ID','SMILES']]
```

```python
df_symp = pd.read_csv('NEW_Table_2B_cid_drug_no_effect.csv')
length=len(df_symp)
```

```python
df_symp
```

```python
drugname=df_symp["drugname"]
side_eff=df_symp["side_effect_name"]
no_side=df_symp["no_of_side_effect"]
```

```python
drugsymp={}
for i in range(length):
    key=drugname[i].lower()
    value=no_side[i]
    drugsymp[key]=value
```

```python
drugsymp
```

```python
df['NoSideEffects'] = df['DRUG_NAME'].str.lower().map(drugsymp)
```

```python
df
```

```python
drugsympall={}

for i in range(length):
    key=drugname[i].lower()
    value=side_eff[i]
    drugsympall[key]=value
```

```python
drugsID={}
CID=df_symp['CID']
drugname=df_symp['drugname']
for i in range(length):
    key=drugname[i].lower()
    value=CID[i]
    drugsID[key]=value
```

```python
df['CID'] = df['DRUG_NAME'].str.lower().map(drugsID)
```

```python
df['AllSideEffects'] = df['DRUG_NAME'].str.lower().map(drugsympall)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(subset = ["NoSideEffects"], inplace=True)
```

```python
df.rename(columns = {'NoSideEffects':'NoSideEffects_Sider'}, inplace = True)
df.rename(columns = {'AllSideEffects':'AllSideEffects_Sider'}, inplace = True)
```

```python
df
```

```python
df.to_csv('NEW_Table_5A_300_Sider_drugmaped.csv', index=False)
```
