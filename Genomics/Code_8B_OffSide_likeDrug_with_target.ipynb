```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
def mapping(df,dfkey,dfvalue):
    key=dfkey
    value=dfvalue
    dictionary={}

    for i in range(len(df)):
        dictionary[key[i]]=value[i]
        
    return dictionary
```

```python
df3=pd.read_csv('NEW_Table_8B_Drug_Target_Sider.csv')
```

```python
Drug1_Target_Name=df3['Drug1_Target_Name']
Drug2_Target_Name=df3['Drug2_Target_Name']

dct={}

for i in tqdm(range(len(df3))):
    value = 0
    if Drug1_Target_Name[i]==Drug2_Target_Name[i]:
        value = 1
    else:
        value = 0
    
    dct[i]=value
```

```python
df3['Matching_Score'] = df3.index.to_series().map(dct)
```

```python
df3=df3.sort_values("Matching_Score", ascending = False)
```
