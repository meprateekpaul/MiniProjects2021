```python
import pandas as pd
import numpy as np
from tqdm import tqdm

# http://db.idrblab.net/ttd/full-data-download
```

```python
df = pd.read_csv('TTD_RAW_P1-07-Drug-TargetMapping.csv')
df
```

```python
df.isnull().sum()
```

```python
df_Offside = pd.read_csv('NEW_Table_5B_300_Offside_drugmaped.csv')
length=len(df_Offside)
df_Offside
```

```python
drugname=df["Drug_Name"]
target_name=df["Target_Name"]
#no_side=df["no_of_side_effect"]
```

```python
drugTarget={}
for i in range(length):
    key=drugname[i].lower()
    value=target_name[i]
    drugTarget[key]=value
```

```python
df_Offside['Target_Name'] = df['Drug_Name'].str.lower().map(drugTarget)
df_Offside
```

```python
drugname=df["Drug_Name"]
Target_ID=df["TargetID"]

drugTargetID={}

for i in range(length):
    key=drugname[i].lower()
    value=Target_ID[i]
    drugTargetID[key]=value
    
df_Offside['TargetID'] = df['Drug_Name'].str.lower().map(drugTargetID)
```

```python
drugname=df["Drug_Name"]
MOA=df["MOA"]
#no_side=df["no_of_side_effect"]

drugTargetID={}

for i in range(length):
    key=drugname[i].lower()
    value=MOA[i]
    drugTargetID[key]=value
    
df_Offside['MOA'] = df['Drug_Name'].str.lower().map(drugTargetID)
```

```python
drugname=df["Drug_Name"]
Highest_status=df["Highest_status"]
#no_side=df["no_of_side_effect"]

drugTargetID={}

for i in range(length):
    key=drugname[i].lower()
    value=Highest_status[i]
    drugTargetID[key]=value
    
df_Offside['Highest_status'] = df['Drug_Name'].str.lower().map(drugTargetID)
```

```python
drugname=df["Drug_Name"]
TTD_DrugID=df["DrugID"]

drugTargetID={}

for i in range(length):
    key=drugname[i].lower()
    value=TTD_DrugID[i]
    drugTargetID[key]=value
    
df_Offside['TTD_DrugID'] = df['Drug_Name'].str.lower().map(drugTargetID)
```
