```python
df_sider=pd.read_csv('NEW_Table_8A_Sider_DrugwithTarget_5nearDrugs_Scored.tsv',)
df_sider
```

```python
def mapping(df,dfkey,dfvalue):
    key=dfkey
    value=dfvalue
    dictionary={}

    for i in range(len(df)):
        dictionary[key[i]]=value[i]
        
    return dictionary
```

```python
s2=mapping(df,df['DRUG_NAME'].str.lower(),df['DRUG_ID'])
```

```python
df_sider['DB_DRUG_ID'] = df_sider['drugname'].str.lower().map(s2)
```

```python
df_sider.dropna(subset = ["DB_DRUG_ID"], inplace=True)
```
