```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
df = pd.read_csv('NEW_Table_3B_DrugScoreMatrix_OffSide.csv')
length=len(df)
field_with_drug_name='drug_concept_name'
```

```python
df
```

```python
df.isnull().sum()
```

```python
dict={}

for i in tqdm(range(length)):
    
    drugnameC=df.columns[5+i]
    df_sort=df.sort_values(by=drugnameC, ascending=False)
    arrangedlist=df_sort[field_with_drug_name].to_numpy()
    
    dict[drugnameC]=arrangedlist[1]+";"+arrangedlist[2]+";"+arrangedlist[3]+";"+arrangedlist[4]+";"+arrangedlist[5]
```

```python
df_final=df_sort[['drug_rxnorn_id',field_with_drug_name,'NoOfSideEffects','SideEffectsOffSide']]
```

```python
df_final['NearDrugs'] = df_final[field_with_drug_name].map(dict)
```

```python
df_final.isnull().sum()
```

```python
df_final.sort_index().to_csv('NEW_Table_7B_OffSide_Drugs_with_NearDrugs.csv', index = False)
```
