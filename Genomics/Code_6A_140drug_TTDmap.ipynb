```python
import pandas as pd
import numpy as np
from tqdm import tqdm

# http://db.idrblab.net/ttd/full-data-download
```

```python
df = pd.read_csv('TTD_RAW_P1-07-Drug-TargetMapping.csv')
df
```

```python
df.isnull().sum()
```

```python
df_140 = pd.read_csv('NEW_Table_5A_300_Sider_drugmaped.csv')
length=len(df_140)
df_140
```

```python
drugname=df["Drug_Name"]
target_name=df["Target_Name"]
#no_side=df["no_of_side_effect"]
```

```python
drugTarget={}
for i in range(length):
    key=drugname[i].lower()
    value=target_name[i]
    drugTarget[key]=value
```

```python
df_140['Target_Name'] = df['Drug_Name'].str.lower().map(drugTarget)
#df_140
```

```python
drugname=df["Drug_Name"]
Target_ID=df["TargetID"]
#no_side=df["no_of_side_effect"]

drugTargetID={}

for i in range(length):
    key=drugname[i].lower()
    value=Target_ID[i]
    drugTargetID[key]=value
    
df_140['TargetID'] = df['Drug_Name'].str.lower().map(drugTargetID)
#df_140
```

```python
drugname=df["Drug_Name"]
MOA=df["MOA"]
#no_side=df["no_of_side_effect"]

drugTargetID={}

for i in range(length):
    key=drugname[i].lower()
    value=MOA[i]
    drugTargetID[key]=value
    
df_140['MOA'] = df['Drug_Name'].str.lower().map(drugTargetID)
#df_140
```

```python
drugname=df["Drug_Name"]
Highest_status=df["Highest_status"]
#no_side=df["no_of_side_effect"]

drugTargetID={}

for i in range(length):
    key=drugname[i].lower()
    value=Highest_status[i]
    drugTargetID[key]=value
    
df_140['Highest_status'] = df['Drug_Name'].str.lower().map(drugTargetID)
#df_140
```

```python
drugname=df["Drug_Name"]
TTD_DrugID=df["DrugID"]
#no_side=df["no_of_side_effect"]

drugTargetID={}

for i in range(length):
    key=drugname[i].lower()
    value=TTD_DrugID[i]
    drugTargetID[key]=value
    
df_140['TTD_DrugID'] = df['Drug_Name'].str.lower().map(drugTargetID)
#df_140
```

```python
#drugname=df["Drug_Name"]
#TTD_DrugID=df["DrugID"]
#no_side=df["no_of_side_effect"]

CID={}
with open("drug_names.tsv") as f:
    for line in f:
        (value, key)=line.split()
        CID[key]=value

df_140['CID_Sidder'] = df_140['DRUG_NAME'].str.lower().map(CID)
#df_140
```
