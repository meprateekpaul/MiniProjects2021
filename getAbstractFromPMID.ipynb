### This script takes the PMID and returns its abstracts from PubMed

#### As sometimes, one needs to read thorough many abstracts from given pmids
#### this script collects it all at one place, for saving searching time

```python
from Bio.Entrez import efetch

def print_abstract(pmid):
    handle = efetch(db='pubmed', id=pmid, retmode='text', rettype='abstract')
    print (handle.read())
```

```python
pmids=[32807018,
32237148,
32617983,
32521062,
32848097,
32848097,
32524611,
32441360,
32702153,
32720703]
```

```python
for ids in pmids:
    print_abstract(ids)
    print("----------------------------------------------------------------")
```
